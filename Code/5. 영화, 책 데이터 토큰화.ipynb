```python
# stopwords 제거 함수
def rem_stop(text):
  new = [word for word in text if word not in stopwords]
  new = [word for word in new if len(word)!= 1]
  return new
```

## 영화 줄거리 토큰화

```python
# 텍스트 클리닝
movie['줄거리_전처리'] = movie['줄거리'].map(lambda x: re.sub('[^\s\w]',' ', x))

# 토크나이징
movie['줄거리_전처리'] = movie['줄거리_전처리'].map(mecab.nouns)
```

```python
movie
```

```python
movie['줄거리_전처리']
```

```python
movie['줄거리_전처리'] = movie['줄거리_전처리'].map(rem_stop)
movie['줄거리_전처리'] = movie['줄거리_전처리'].map(lambda x: ' '.join(x))
```

```python
# 키워드 ',' 단위로 구분
def keyword(text):
  if type(text) == str:
    text = text.split(',')
  else:
    text = ['']
  return text
```

```python
movie['키워드'] = movie['키워드'].map(keyword)
movie['키워드'] = movie['키워드'].map(lambda x: ' '.join(x))
```

```python
movie
```

```python
movie['줄거리+키워드'] = movie['줄거리_전처리'] + movie['키워드']
movie
```

```python
# 영화명에 띄어쓰기 없애기
def title(text):
  text = str(text)
  text = text.replace(' ','')
  return text

movie['영화명_전처리'] = movie['영화명'].map(title)
```
