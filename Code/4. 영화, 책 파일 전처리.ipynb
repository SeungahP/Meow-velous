```python
# 시리즈물/에디션/특별판 책 평점 처리
dup_book = list(kyobo_list[kyobo_list.duplicated(subset=['user','book'])]['book'].unique())
book_list2 = [book for book in dup_book if book not in book_list1]
rating_book2 = pd.DataFrame()

for title in book_list2:
    df = kyobo_list[kyobo_list['book']==title]
    user = list(df[df.duplicated(subset='user')]['user'].values)
    user_df1 = df[df['user'].map(lambda x: x in user)]
    user_df2 = df[df['user'].map(lambda x: x not in user)]
    rating_avg = user_df1.groupby('user').mean()
    rating_avg = rating_avg.rename(columns={'rating':'rating_avg'})
    rating_avg.reset_index(inplace=True)
    new_df1 = pd.merge(user_df1, rating_avg)
    new_df1.drop(['rating'], axis=1, inplace=True)
    new_df1 = new_df1.rename(columns={'rating_avg':'rating'})
    new_df1.drop_duplicates(subset='user', inplace=True)
    new_df = pd.concat([new_df1, user_df2])
    rating_book2 = pd.concat([rating_book2, new_df])
```
